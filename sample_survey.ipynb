### Sample survey analysis

This notebook does some simple analysis on the survey result.

```python
import pandas as pd

# Read in the csv file
data_url = 'sample_survey.csv'
# Reading the data
df = pd.read_csv(data_url)
df
```

```python
column_name = 'Name something you never want your kid to bring home.'
df[column_name].value_counts(sort=True)
```

```python
column_name = 'Name something you never want your kid to bring home.'
df[column_name].str.strip().str.lower().value_counts(sort=True)
```

```python
df[column_name].describe()
```

```python
for column in df:
      
      # process each column
      print('======================================================================')
      print('Column Name : ', column)
      print('======================================================================')
      print('====== column statistics: ======')
      print(df[column].describe())
      print('====== column value counts: ======')
      print(df[column].str.strip().str.lower().value_counts(sort=True))
```

```python
# write results to file
import sys
original_stdout = sys.stdout # Save a reference to the original standard output

with open('sample_survey_result.txt', 'w') as f:
  sys.stdout = f # Change the standard output to the file we created.

  # Iterate over each column
  for column in df:
      
      # process each column
      print('======================================================================')
      print('Column Name : ', column)
      print('======================================================================')
      print('====== column statistics: ======')
      print(df[column].describe())
      print('====== column value counts: ======')
      print(df[column].str.strip().str.lower().value_counts(sort=True))

  # close the file
  f.close()

# change the standard output back
sys.stdout = original_stdout # Reset the standard output to its original value
```
